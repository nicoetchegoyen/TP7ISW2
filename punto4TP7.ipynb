```python
import numpy as np
import matplotlib.pyplot as plt
```

# Definir funciones para calcular el esfuerzo y el tiempo calendario


```python
def calcular_esfuerzo(S):
    return 8 * S**0.95


def calcular_tiempo(E):
    return 2.4 * E**0.33
```

# Crear intervalos de tamaños y esfuerzos

```python
tamaños = np.linspace(0, 10000, 500) 
esfuerzos = calcular_esfuerzo(tamaños)
```

# Intervalo para los tiempos en función del esfuerzo [1, 500]

```python
esfuerzos_intervalo = np.linspace(1, 500, 500)
tiempos = calcular_tiempo(esfuerzos_intervalo)
```

# Graficar esfuerzo vs tamaño del proyecto

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(tamaños, esfuerzos, label='Esfuerzo E(S)', color='blue')
plt.title('Esfuerzo en función del tamaño del proyecto')
plt.xlabel('Tamaño del Proyecto (S)')
plt.ylabel('Esfuerzo (E)')
plt.grid(True)
```

# Graficar tiempo calendario vs esfuerzo

```python
plt.subplot(1, 2, 2)
plt.plot(esfuerzos_intervalo, tiempos, label='Tiempo Td(E)', color='green')
plt.title('Tiempo calendario en función del esfuerzo')
plt.xlabel('Esfuerzo (E)')
plt.ylabel('Tiempo Calendario (Td)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
